```python
import pandas as pd
from  matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#Read Data
data = pd.read_csv("c:\\Waste_type.csv")
data
```

```python
#Sorting

sorted = data.sort_values("year")
sorted
```

```python
data_1960 = sorted.head(4)
data_1960 =data_1960.sort_values("type_id")
data_1960
```

```python
labels_1 = "recycled","combusted","landfilled"
labels_1
```

```python
plt.pie(data_1960.plastic[1:4,],labels = labels_1,autopct = '%1.1f%%')

plt.suptitle("Plastic Waste Disposal Distribution(1960)",fontsize=15)
plt.title("Total  Plastic Waste Generated = 39(Million Tonnes)")
```
